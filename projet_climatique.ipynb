```python
import csv

chemin_fichier = "2024.csv"
donnees = {}

with open(chemin_fichier, newline='', encoding='utf-8-sig') as fichier:
    lecteur = csv.DictReader(fichier, delimiter=';')
    
    colonnes_valides = []
    for colonne in lecteur.fieldnames:
        if colonne and colonne.strip() != '':
            colonnes_valides.append(colonne)

    for nom_colonne in colonnes_valides:
        donnees[nom_colonne] = []

    colonnes_int = ['Annee', 'Mois', 'Jour']
    colonnes_float = ['Temperature_maximum', 'Temperature_minimum', 'Temperature_moyenne', 'Precipitations']

    for ligne in lecteur:
        for nom_colonne in colonnes_valides:
            valeur = ligne[nom_colonne]
            if nom_colonne in colonnes_int:
                donnees[nom_colonne].append(int(valeur))
            elif nom_colonne in colonnes_float:
                donnees[nom_colonne].append(float(valeur))
            else:
                donnees[nom_colonne].append(valeur)

print(donnees)
```
